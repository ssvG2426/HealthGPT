```python
# ===================== CELL 2: IMPORTS & SETUP (DRUG INFO) =====================
import warnings
warnings.filterwarnings("ignore")

import re
import json
import time
import pickle
from pathlib import Path
from dataclasses import dataclass
from typing import List, Dict, Tuple

import numpy as np
import torch
import faiss
from datasets import load_dataset
from sentence_transformers import SentenceTransformer
from nltk.tokenize import word_tokenize, sent_tokenize
from rank_bm25 import BM25Okapi

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"ðŸ”§ Using device: {device}")

@dataclass
class DomainConfig:
    name: str = "drug_info"
    dataset_name: str = "lavita/ChatDoctor-HealthCareMagic-100k"
    dataset_split: str = "train"
    chunk_window: int = 3
    chunk_stride: int = 1
    embed_model: str = "sentence-transformers/all-MiniLM-L6-v2"

config = DomainConfig()

print(f"ðŸ“‹ Building index for: {config.name}")
print(f"ðŸ“¦ Dataset: {config.dataset_name}")
```

```python
# ===================== CELL 3: DATASET LOADING & FILTERING (DRUG INFO) =====================

# Drug/Medicine keywords for filtering
DRUG_KEYWORDS = [
    "drug", "medicine", "medication", "tablet", "capsule", "pill", "dose", "dosing", "dosage",
    "side effect", "adverse effect", "contraindication", "antibiotic", "painkiller", "analgesic",
    "paracetamol", "aspirin", "acetaminophen", "ibuprofen", "prescription", "interaction",
    "antidepressant", "antihypertensive", "steroid", "chemotherapy", "brand", "generic",
    "pharmacy", "pharmacology", "therapeutic", "treatment", "remedy", "over-the-counter",
    "OTC", "rx", "pharmaceutical", "antihistamine", "anticoagulant", "insulin", "statin"
]

def is_drug_topic(q: str, a: str):
    """Check if Q&A contains drug/medicine-related keywords."""
    text = (q + " " + a).lower()
    return any(kw in text for kw in DRUG_KEYWORDS)

def extract_drug_qa(dataset) -> list:
    """Extract only drug-related Q&A pairs."""
    qa_data = []
    for idx, row in enumerate(dataset):
        try:
            question = row.get("instruction", "") or row.get("question", "")
            answer = row.get("output", "") or row.get("answer", "")
            
            if question and answer and len(answer) > 20 and is_drug_topic(question, answer):
                qa_data.append({
                    "question": str(question).strip(),
                    "answer": str(answer).strip(),
                    "source_id": idx
                })
        except Exception as e:
            if idx < 3:
                print(f"âš ï¸ Skipped row {idx}: {e}")
    return qa_data

# Load dataset
print(f"ðŸ“¥ Loading {config.dataset_name}...")
dataset = load_dataset(config.dataset_name, split=config.dataset_split)
print(f"âœ… Loaded {len(dataset)} rows")

# Extract drug-related Q&A
print(f"\nðŸ” Filtering for drug/medicine Q&A pairs...")
qa_data = extract_drug_qa(dataset)
print(f"âœ… Extracted {len(qa_data)} drug/medicine Q&A pairs")

if len(qa_data) < 100:
    print(f"âš ï¸ Only {len(qa_data)} drug Q&A pairs extracted! Keywords may be too strict.")
else:
    print(f"\nðŸ“ Sample drug Q: {qa_data[0]['question'][:120]}...")
    print(f"   Sample drug A: {qa_data[0]['answer'][:120]}...")
```

```python
# ===================== CELL 4: TEXT CHUNKING =====================
def create_chunks(data: List[Dict], window: int = 3, stride: int = 1, min_chars: int = 10) -> List[Dict]:
    chunks = []
    for item in data:
        text = item.get("answer", "")
        if not text or len(text) < min_chars:
            continue
        sentences = sent_tokenize(text)
        if not sentences:
            continue
        if len(sentences) <= window:
            chunks.append({
                "chunk": " ".join(sentences),
                "source_idx": item.get("source_id", -1),
                "chunk_id": len(chunks)
            })
            continue
        for i in range(0, max(1, len(sentences) - window + 1), stride):
            chunks.append({
                "chunk": " ".join(sentences[i:i + window]),
                "source_idx": item.get("source_id", -1),
                "chunk_id": len(chunks)
            })
    return chunks

print(f"ðŸ”ª Creating chunks (window={config.chunk_window}, stride={config.chunk_stride}, min_chars=10)...")
chunks = create_chunks(qa_data, window=config.chunk_window, stride=config.chunk_stride, min_chars=10)
print(f"âœ… Created {len(chunks)} chunks")

if len(chunks) == 0:
    print("[âŒ ERROR] No chunks created! Check QA extraction or chunking settings.")
else:
    print("\nðŸ“ Sample chunks:")
    for i, chunk in enumerate(chunks[:3]):
        print(f"{i+1}. {chunk['chunk'][:120]}...")
```

```python
# ===================== CELL 5: BUILD FAISS INDEX =====================
if len(chunks) == 0:
    raise ValueError("âŒ ERROR: No chunks created! Check previous cell.")

print(f"ðŸ“¦ Loading embedder: {config.embed_model}")
embedder = SentenceTransformer(config.embed_model, device=device)
print(f"âœ… Embedder loaded")

id2doc = [chunk["chunk"] for chunk in chunks]
print(f"ðŸ“Š Encoding {len(id2doc)} chunks...")

if len(id2doc) == 0:
    raise ValueError("âŒ ERROR: id2doc is empty! No valid chunks to encode.")

embeddings = embedder.encode(
    id2doc,
    normalize_embeddings=True,
    show_progress_bar=True,
    batch_size=64,
    convert_to_numpy=True
).astype('float32')

print(f"âœ… Embeddings shape: {embeddings.shape}")

if len(embeddings.shape) != 2:
    raise ValueError(f"âŒ ERROR: Invalid embeddings shape {embeddings.shape}. Expected 2D array.")

dim = embeddings.shape[1]
index = faiss.IndexFlatIP(dim)
index.add(embeddings)

print(f"âœ… FAISS index built: {index.ntotal} vectors, dimension {dim}")
```

```python
# ===================== CELL 6: BUILD BM25 INDEX =====================
print(f"ðŸ”¨ Building BM25 index...")
bm25_corpus = [word_tokenize(doc.lower()) for doc in id2doc]
bm25 = BM25Okapi(bm25_corpus)
print(f"âœ… BM25 index built")
```

```python
# ===================== CELL 7: SAVE INDEXES =====================
index_file = f"{config.name}_faiss.index"
id2doc_file = f"{config.name}_id2doc.pkl"
metadata_file = f"{config.name}_metadata.json"

print(f"ðŸ’¾ Saving FAISS index to {index_file}...")
faiss.write_index(index, index_file)

print(f"ðŸ’¾ Saving id2doc to {id2doc_file}...")
with open(id2doc_file, "wb") as f:
    pickle.dump(id2doc, f)

metadata = {
    "created_at": time.time(),
    "domain": config.name,
    "dataset_path": config.dataset_name,
    "n_vectors": int(index.ntotal),
    "embedding_dim": dim,
    "chunk_window": config.chunk_window,
    "chunk_stride": config.chunk_stride,
    "embed_model": config.embed_model
}

print(f"ðŸ’¾ Saving metadata to {metadata_file}...")
with open(metadata_file, "w") as f:
    json.dump(metadata, f, indent=2)

print("\n" + "="*80)
print("âœ… INDEX BUILDING COMPLETE!")
print("="*80)
print(f"ðŸ“ Files created:")
print(f"  - {index_file}")
print(f"  - {id2doc_file}")
print(f"  - {metadata_file}")
print(f"\nðŸ“Š Statistics:")
print(f"  - Total vectors: {index.ntotal}")
print(f"  - Embedding dimension: {dim}")
print(f"  - Source Q&A pairs: {len(qa_data)}")
print(f"  - Chunks created: {len(chunks)}")
print("="*80)
```

```python
# ===================== CELL 8: VERIFY INDEX (OPTIONAL) =====================
def test_retrieval(query: str, top_k: int = 5):
    print(f"\nðŸ” Testing query: '{query}'")
    query_emb = embedder.encode([query], normalize_embeddings=True, convert_to_numpy=True).astype('float32')
    D, I = index.search(query_emb, top_k)
    print(f"\nðŸ“‹ Top {top_k} results:")
    for i, (idx, score) in enumerate(zip(I[0], D[0]), 1):
        print(f"\n{i}. Score: {score:.4f}")
        print(f"   {id2doc[idx][:200]}...")

test_queries = [
    "What are the side effects of ibuprofen?",
    "Can I take aspirin and warfarin together?",
    "What is the correct dosage of paracetamol for adults?",
    "What medications are used to treat high blood pressure?",
    "Are there any contraindications for metformin?"
]

print("\n" + "="*80)
print("ðŸ§ª TESTING DRUG INFO INDEX")
print("="*80)

for query in test_queries:
    test_retrieval(query, top_k=3)
```
